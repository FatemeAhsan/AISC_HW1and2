```python
df_nona_no_date = df_nona.drop('date', axis=1)
df_nona_no_date
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
encoder = LabelEncoder()

df_encoded = df_nona_no_date.copy()

df_encoded['quarter'] = encoder.fit_transform(df_encoded['quarter'])
df_encoded['department'] = encoder.fit_transform(df_encoded['department'])
```

```python
for c in df_encoded.columns[:-1]:
  plt.figure(figsize=[4, 3])
  plt.scatter(df_encoded[[c]], df_encoded[df_encoded.columns[-1]], color=np.random.rand(3))
  plt.xlabel(c)
  plt.ylabel(df.columns[-1])
  plt.title(f'{df.columns[-1]} with {c}')
  plt.show()
```

```python
df_encoded_one_hot = df_encoded.drop('quarter', axis=1)
df_encoded_one_hot = df_encoded_one_hot.join(pd.get_dummies(df_encoded['quarter'], prefix='q'))
df_encoded_one_hot
```

```python
df_y = df_encoded_one_hot[['actual_productivity']]
df_x = df_encoded_one_hot.drop('actual_productivity', axis=1)
df_x
```

```python
df_y
```

```python
from sklearn.model_selection import train_test_split
```

```python
df_x_train, df_x_val, df_y_train, df_y_val = train_test_split(df_x, df_y, test_size=0.3, random_state=42)

df_x_val, df_x_test, df_y_val, df_y_test = train_test_split(df_x_val, df_y_val, test_size=0.45, random_state=42)

df_x_train.shape, df_y_train.shape, df_x_val.shape, df_y_val.shape, df_x_test.shape, df_y_test.shape,
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()

x_train_nrm = scaler.fit_transform(df_x_train)
x_val_nrm = scaler.transform(df_x_val)
x_test_nrm = scaler.transform(df_x_test)
```

```python
x_train_nrm
```

```python
x_train_nrm.shape
```

```python
y_train = df_y_train.values
y_val = df_y_val.values
y_test = df_y_test.values
```

```python
from keras.models import Sequential
from keras.layers import Dense
```

```python
model1 = Sequential()

model1.add(Dense(256, activation='relu', input_shape=x_train_nrm.shape[1:]))

model1.add(Dense(64, activation='relu'))

model1.add(Dense(32, activation='relu'))

model1.add(Dense(1, 'linear'))

model1.compile(optimizer='adam', loss='mse', metrics='mae')

model1.summary()
```

```python
import tensorflow as tf
```

```python
results = model1.fit(x_train_nrm, 
                    y_train, 
                    epochs=20,
                    batch_size=32,
                    validation_data=(x_val_nrm, y_val),
                    callbacks=[tf.keras.callbacks.EarlyStopping(monitor='val_loss', patience=5, restore_best_weights=True, verbose=1)])
```

```python
model1.evaluate(x_test_nrm, y_test)
```

```python
from sklearn.metrics import r2_score
```

```python
y_test_hat = model1.predict(x_test_nrm)
```

```python
r2_score(y_test, y_test_hat)
```

```python
def plot_losses():
    plt.plot(results.history['loss'], label='loss')
    plt.plot(results.history['val_loss'], label='val_loss')
    plt.legend()
```

```python
plot_losses()
```
