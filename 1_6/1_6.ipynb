```python
import pandas as pd
```

```python
df = pd.read_csv('files/possum.csv').drop('case', axis=1)
df
```

```python
df_nona = df.dropna()
df_nona
```

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
for c in df_nona.columns:
    plt.figure(figsize=[3, 2])
    plt.title(c)
    plt.hist(df_nona[c], color=np.random.rand(3))
    plt.show()
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
encoder = LabelEncoder()

df_encoded = df_nona.copy()

df_encoded['Pop'] = encoder.fit_transform(df_encoded['Pop'])
df_encoded['sex'] = encoder.fit_transform(df_encoded['sex'])
```

```python
df_encoded
```

```python
df_encoded0 = df_encoded[df_encoded['sex'] == 0]
df_encoded1 = df_encoded[df_encoded['sex'] == 1]
df_encoded0.shape, df_encoded1.shape 
```

```python
df_encoded1_e0 = df_encoded1.sample(df_encoded0.shape[0])
df_encoded0.shape, df_encoded1_e0.shape 
```

```python
df_encoded_e = pd.concat([df_encoded0, df_encoded1_e0])
df_encoded_e.shape
```

```python
df_y = df_encoded_e[['sex']]
df_x = df_encoded_e.drop('sex', axis=1)
df_x
```

```python
from sklearn.model_selection import train_test_split
```

```python
df_x_train, df_x_val, df_y_train, df_y_val = train_test_split(df_x, df_y, test_size=0.15)

df_x_train.shape, df_y_train.shape, df_x_val.shape, df_y_val.shape
```

```python
df_x_train
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()

x_train_nrm = scaler.fit_transform(df_x_train)
x_val_nrm = scaler.transform(df_x_val)
```

```python
x_train_nrm
```

```python
x_train_nrm.shape
```

```python
from keras.models import Sequential
from keras.layers import Dense
```

```python
model1 = Sequential()

model1.add(Dense(256, activation='relu', input_shape=x_train_nrm.shape[1:]))

model1.add(Dense(64, activation='relu'))

model1.add(Dense(32, activation='relu'))

model1.add(Dense(1, 'sigmoid'))

model1.compile(optimizer='adam', loss='binary_crossentropy', metrics='acc')

model1.summary()
```

```python
y_train = df_y_train.values
y_val = df_y_val.values
```

```python
import tensorflow as tf
```

```python
results = model1.fit(x_train_nrm, 
                    y_train, 
                    epochs=20,
                    batch_size=32,
                    validation_data=(x_val_nrm, y_val),
                    callbacks=[tf.keras.callbacks.EarlyStopping(monitor='val_loss', patience=5, restore_best_weights=True, verbose=1)])
```

```python
def plot_losses():
    plt.plot(results.history['loss'], label='loss')
    plt.plot(results.history['val_loss'], label='val_loss')
    plt.legend()
```

```python
plot_losses()
```
